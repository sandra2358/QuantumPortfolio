# Shor's Algorithm

Shor's Algorithm is one of the most famous quantum algorithm developed by Peter Shor in 1994. It is capable of factoring large numbers into prime numbers in polynominal time. With a sufficiently power quantum computer, it can break the RSA encryption protocol that is widely employed. 

Here, we focus on the quantum part of the Shor's algorithm, which primarily solves the period finding problem. The problem of factoring a large number into two prime numbers can be turned into a period finding problem.

## 1. Period Finding 

Before diving into Shor's Algorithm, let's first define and solve the period finding problem. The period finding function is: $$f(x)=a^x \; \text{mod}\; N$$
where mod is the modulo operation, and $a$ and $N$ are coprime integers, meaning the greatest common factor of $a$ and $N$ ($gcd(a,N)$) is 1. 

#### The modulo operation 
>The modulo operation, $x\;\text{mod}\;y$, finds the remainder when dividing $x$ by $y$. For example, when we want to find the remainder of 21 when divided by 4, we can express it as: $$ 21 \;\text{mod}\;4 = 1$$
>
>We can represent it in another method: 
$$ 21 = 1 \;\; (\text{mod} \;4) $$

We can find the period of $f$ by finding the smallest $r$ which $$ f(x+r)= f(r) $$ and $r$ is the period.  

To verify, $r$ should satisfy the condition: $ f(r)= a^r \;\text{mod} \; N =1 $

For simplicity, we can set $x=0$, thus we would be finding the first $r$ that satisifies: $$ f(r)= a^r \;\text{mod} \; N =1 $$ 


Below, we look at the case where we set $N=21$ and $a=4$. 

```python
N = 21
a = 4


# Calculate the values of f(x) given the above a and N
x_range = 30
x_values = np.arange(x_range)
y_values = [np.mod(a**x, N) for x in x_values]

# Display the data using matplotlib
fig, ax = plt.subplots()
ax.plot(x_values, y_values, linewidth=1, linestyle='dotted', marker='x')
ax.set(xlabel='$x$', ylabel=f'${a}^x$ mod ${N}$',
       title="Example of Periodic Function")

# Attempting to find the range r 
try: 
    print("The period is r =",y_values[1:].index(1)+1)
except ValueError:
    print("Cannot find period. Check that a should be smaller than N, and that a and N are coprime numbers. Otherwise, increase the x_range. ")
except: 
    print("Error")
```

For simplicity, we define the function *findPeriod* to find the period when given $a$ and $N$. Let's use this function to find the period where $a = 3$ and $N = 1177$. 

```python
# Input: a: integer
#        N: integer
# Output: r: integer
# If the output is 0, then attempt with a different a

def findPeriod(a,N,x_range=30):
    # initialise the variable r, representing the period 
    r = 0  

    count = 0 
    
    # while loop to ensure the range is found if it is numerically possible to find 
    # adding the count is to prevent looping indefinitely 
    while r == 0 and count <5:
        
        # Calculate the values of f(x) given the above a and N
        x_values = np.arange(x_range)
        y_values = [np.mod(a**x, N) for x in x_values]

        # Attempting to find the range r 
        try: 
            r=y_values[1:].index(1)+1
        except ValueError:
            x_range *=5
            r=0 
            count +=1 
    
    return r
```

```python
# Finding the period where a = 3 and N = 1177
print(findPeriod(3,1177))
```
